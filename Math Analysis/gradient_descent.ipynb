```python
def grad_descent(x0, lr, n):
    x = x0
    
    x_list = []
    f_list = []
    for i in range(n):
        grad = calc_grad(x)
        x = x - lr * grad
        x_list.append(x)
        f_list.append(f(x))
        # print(f'Iteration: {i}, x: {x}, f(x): {f(x)}, grad(x): {grad}')
    return x, x_list, f_list
```

```python
# f(x, y) = x^2 - 10 * cos(2pi*x) + 10 + y^2 - 10 * cos(2pi*y) + 10



def f(x, y):
    return x**2 - 10 * np.cos(2 * np.pi * x) + 1 + y**2 - 10 * np.cos(2 * np.pi * y) + 10

def calc_grad(x, y):
    x_grad = 2 * x + 20 * np.pi * np.sin(2 * np.pi * x)
    y_grad = 2 * y + 20 * np.pi * np.sin(2 * np.pi * y)
    
    return x_grad, y_grad
```

```python
def grad_descent_2d(x0, y0, lr, n):
    x = x0
    y = y0
    
    x_list = []
    y_list = []
    f_list = []
    
    for i in range(n):
        x_grad, y_grad = calc_grad(x, y)
        x = x - lr * x_grad
        y = y - lr * y_grad
        
        x_list.append(x)
        y_list.append(y)
        f_list.append(f(x, y))
        # print(f'Iteration: {i}, x: {x}, y: {y}, f(x): {f(x, y)}, grad(x): {x_grad}, grad(y): {y_grad}')
    return x, y, x_list, y_list, f_list
```

```python
x, y, x_list, y_list, f_list = grad_descent_2d(-1, -1, 0.005, 50)
```

```python
def plot(x_values, y_values, f_values, min_value=-3, max_value=3):
    x_true = np.linspace(min_value, max_value, 500)
    y_true = np.linspace(min_value, max_value, 500)
    x_true, y_true = np.meshgrid(x_true, y_true)
    f_true = f(x_true, y_true)
    
    plt.contour(x_true, y_true, f_true, levels=50)
    
    plt.scatter(x_values, y_values, color='red')
    plt.plot(x_values, y_values, color='red', linestyle='--')
    
    
    plt.title('Градиентный спуск')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
x, y, x_list, y_list, f_list = grad_descent_2d(-1.3, -1.2, 0.009, 100)
plot(x_list, y_list, f_list, min_value=-1.5, max_value=1.5)
```

```python
def grad_descent_2d(x0, y0, lr, n, tol=0.0001):
    x = x0
    y = y0
    
    x_list = []
    y_list = []
    f_list = []
    
    f_prev = np.inf
    
    for i in range(n):
        x_grad, y_grad = calc_grad(x, y)
        x = x - lr * x_grad
        y = y - lr * y_grad
        
        f = f(x, y)
        
        if abs(f - f_prev) < tol:
            break
        
        f_prev = f
        
        x_list.append(x)
        y_list.append(y)
        f_list.append()
        # print(f'Iteration: {i}, x: {x}, y: {y}, f(x): {f(x, y)}, grad(x): {x_grad}, grad(y): {y_grad}')
    return x, y, x_list, y_list, f_list
```
